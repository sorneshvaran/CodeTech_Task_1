# Task 1: USE PYTHON TO FETCH DATA FROM A PUBLIC API AND CREATE VISUALIZATIONS USING MATPLOTLIB OR SEABORN.

## 1. Fetching Data from public api ==> Agify.io

```python
import requests

#Using Agify api to fetch data => Storing different names to get more than one data
names = ['Einstein','Arya','Jack','Raman','Newton']
age = []

#Creating loop to get the data for all the names
for i in names: 
    #Stroing the api in url variable
    url = f"https://api.agify.io?name={i}"
    
    #Getting the data using .get() method
    response = requests.get(url)
    
    #Coverting json object to python dictionary 
    data = response.json()

    #Appending age for data visualization
    age.append(data['age'])
    
    print(f"Name:{data['name']} => Age:{data['age']}")
```

## 2. Visulization: Predicted Age by Name Using Agify API

```python
import matplotlib.pyplot as plt
```

```python
#Creating a barchat to show Name to Age visual representation
plt.bar(x = names, height = age)
plt.title("Predicted Age by Name Using Agify API")
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()
```
